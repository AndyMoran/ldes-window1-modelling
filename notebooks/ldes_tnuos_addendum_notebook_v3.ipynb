## Storage-specific tariff calculation

Formula: `Peak + ALF × (Year Round Shared + Year Round Not Shared) + generation adjustment`.


```python
peak = 3.62806
yr_shared = 26.532717
yr_not_shared = 17.674559
generation_adjustment = -2.47676
battery_alf = 0.080542
psh_alf = 0.109859

battery_tariff = peak + battery_alf * (yr_shared + yr_not_shared) + generation_adjustment
psh_tariff = peak + psh_alf * (yr_shared + yr_not_shared) + generation_adjustment

pd.DataFrame([
    {"storage_type": "Battery storage", "generic_alf": battery_alf, "generation_tariff_gbp_kwyr": battery_tariff},
    {"storage_type": "Pumped storage", "generic_alf": psh_alf, "generation_tariff_gbp_kwyr": psh_tariff},
])
```
